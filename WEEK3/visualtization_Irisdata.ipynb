### *Importing Libraries and Dataset*

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df = pd.read_csv("Iris.csv")
```

```python
df.head()
```

### *Visualization*
**1. Line Plot**

```python
plt.plot(df['Species'])
plt.xlabel('Count')
plt.ylabel('Species')
plt.title('Line Plot')
plt.show()
```

```python
plt.figure(figsize=(8, 6))
sns.lineplot(data=df,x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Line Plot : Sepal Length vs Sepal Width')
plt.show()
```

---
**2. Box Plot**

```python
plt.boxplot(df['PetalLengthCm'])
plt.title('Box Plot')
plt.show()
```

```python
sns.boxplot(data=df, x='Species', y='SepalLengthCm')
plt.title('Box Plot : Sepal Length by Species')
plt.show()
```

---
**3. Histogram**

```python
sns.histplot(data=df, x='PetalLengthCm', hue='Species', kde=True)
plt.xlabel('PetalLengthCm')
plt.ylabel('Count')
plt.title('Histogram')
plt.show()
```

```python
sns.histplot(data=df, x='SepalLengthCm', hue='Species', kde=True)
plt.title('Histogram : Sepal Length')
plt.show()
```

---
**4. Scatter Plot**

```python
plt.scatter(x=df['SepalLengthCm'],y=df['PetalLengthCm'])
plt.xlabel('SepalLengthCm')
plt.ylabel('PetalLengthCm')
plt.title('Scatter Plot')
```

```python
sns.scatterplot(data=df, x='SepalLengthCm', y='SepalWidthCm', hue='Species')
plt.title('Scatter Plot : Sepal Length vs Sepal Width')
plt.show()
```

---
**5. Bar Graph**

```python
plt.figure(figsize=(8, 6))
df.groupby('Species').mean().plot(kind='bar')
plt.title('Bar Plot : Mean Measurements by Species')
plt.ylabel('Mean Value')
plt.show()
```

---
**6. Pair Plot**

```python
sns.pairplot(df, hue='Species')
plt.suptitle("Pair Plot of Iris Dataset", y=1.02)
plt.show()
```

---
**7. 3D Scatter Plot**

```python
ax = plt.axes(projection='3d')
ax.scatter(df.index,df['SepalLengthCm'],df['PetalLengthCm'])
ax.set_xlabel('Index')
ax.set_ylabel('SepalLengthCm')
ax.set_zlabel('PetalLengthCm')
plt.show()
```

---
**8. Heat Map**

```python
from sklearn.preprocessing import LabelEncoder
label_encoder = LabelEncoder()
df['Species'] = label_encoder.fit_transform(df['Species'])
```

```python
sns.heatmap(df.corr(), annot=True, cmap='coolwarm')
plt.title('Heatmap : Correlation Matrix')
plt.show()
```
